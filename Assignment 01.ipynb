## Assignment 1 on Intensity Transformations and Neighborhood Filtering

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
c=np.array([[0,0],[50,50],[50,100],[150,255],[150,150],[255,255]])

seg1=np.linspace(c[0,1],c[1,1], c[1,0]-c[0,0]+1).astype('uint8')
print(seg1)
```

```python
seg2=np.linspace(c[2,1]+1,c[3,1],c[3,0]-c[2,0]).astype('uint8')
print(seg2)
```

```python
seg3=np.linspace(c[4,1]+1,c[5,1],c[5,0]-c[4,0]).astype('uint8')
print(seg3)
```

```python
tranfrom=np.concatenate((seg1, seg2, seg3),axis=0).astype('uint8')
print(tranfrom)
```

```python
print(len(tranfrom  ))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.arange(len(tranfrom)), tranfrom, marker='o')
plt.title('Intensity Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.grid()
plt.xlim(0, 260)
plt.ylim(0, 260)
plt.show()
```

```python
img = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
```

```python
img_transformed = cv.LUT(img, tranfrom)
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_transformed, cmap='gray')
plt.title('Transformed Image')
plt.axis('off')
plt.show()

```

## Q(2)
Apply a similar operation as above (question 1) to accentuate
(a) white matter
(b) gray matter
in the brain proton density image shown in Fig. 2. Show the intensity transformations as a plots.


### White Matter


```python
thereshold1 = 190
```

```python
seg1=np.linspace(0,0, thereshold1+1).astype('uint8')
seg2=np.linspace(thereshold1+1,255, 255-thereshold1).astype('uint8')
#seg3=np.linspace(255,255, 255-201+1).astype('uint8')
tranfrom2=np.concatenate((seg1, seg2),axis=0).astype('uint8')
print(tranfrom2)
```

```python
print(len(tranfrom2))
```

```python
img1 = cv.imread('brain.png',cv.IMREAD_GRAYSCALE)
img_transformed2 = cv.LUT(img1, tranfrom2)
```

```python

plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.plot(np.arange(len(tranfrom2)), tranfrom2, marker='o')
plt.title('Intensity Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.grid()
plt.xlim(0, 260)
plt.ylim(0, 260)
plt.subplot(1, 3, 2)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(img_transformed2, cmap='gray')
plt.title('Transformed Image- White Matter')
plt.axis('off')
plt.show()
```

### Gray Matter

```python
thereshold1 = 125
thereshold2 = 180
```

```python
seg1=np.linspace(0,0, thereshold1+1).astype('uint8')
seg2=np.linspace(thereshold1+1,thereshold2, thereshold2-thereshold1).astype('uint8')
seg3=np.linspace(0,0, 255-thereshold2).astype('uint8')

tranfrom3=np.concatenate((seg1, seg2, seg3),axis=0).astype('uint8')
print(len(tranfrom3))
```

```python
img_transformed3=cv.LUT(img1, tranfrom3)
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.plot(np.arange(len(tranfrom3)), tranfrom3, marker='o')
plt.title('Intensity Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.grid()
plt.xlim(0, 260)
plt.ylim(0, 260)
plt.subplot(1, 3, 2)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(img_transformed3, cmap='gray')
plt.title('Transformed Image- Gray Matter')
plt.axis('off')
plt.show()
```
